```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

buttons = cv.imread("./img/set-of-colorful-buttons-vector.jpg")
buttons = cv.cvtColor(buttons, cv.COLOR_BGR2RGB)

plt.imshow(buttons)
```

```python
light_purp = (233, 187, 240)
dark_purp = (94, 4, 107)
purp_mask = cv.inRange(buttons, dark_purp, light_purp)

output = cv.bitwise_and(buttons, buttons, mask=purp_mask)

plt.imshow(output)
```

```python

boundaries = (([17, 15, 100], [50, 56, 200]),
	([86, 31, 4], [220, 88, 50]),
	([25, 146, 190], [62, 174, 250]),
	([103, 86, 65], [145, 133, 128]))

for i in range(4):
    bgr_but = cv.imread("./img/set-of-colorful-buttons-vector.jpg")

    light = np.array(boundaries[i][1])
    dark = np.array(boundaries[i][0])
    mask = cv.inRange(bgr_but, dark, light)

    contours, _ = cv.findContours(mask, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    output = cv.drawContours(bgr_but, contours, -1, (0, 255, 0), 3)
    e = cv.cvtColor(output, cv.COLOR_BGR2RGB)
    plt.figure()
    plt.imshow(e)
```
